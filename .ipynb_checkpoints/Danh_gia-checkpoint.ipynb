```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.DataFrame({'groups': ['k_NN','k_NN',
                                'Linear_Regression','Linear_Regression',
                                'Ridge_Regression', 'Ridge_Regression',
                                'Lasso_Regression','Lasso_Regression',
                                'Decision_Tree','Decision_Tree',
                                'Random_Forest', 'Random_Forest',
                                'Neural_Network', 'Neural_Network'], 
                     'type': ['MSE_Train','MSE_Test','MSE_Train','MSE_Test','MSE_Train','MSE_Test','MSE_Train','MSE_Test',
                              'MSE_Train','MSE_Test','MSE_Train','MSE_Test','MSE_Train','MSE_Test'],
                     'value': [135259,139028,
                               517659,91823,
                               513183,109500,
                               550842,148812,
                               136980,216426,
                               80370,153730,
                               797153,325785]})

pivoted = data.pivot(index='groups', columns='type', values='value')

ax = pivoted.plot.bar(stacked=False, color=['b','g'])
ax.set_title('Model selection dùng phương pháp holdout')
ax.set_xticklabels(['k_NN', 'Linear_Regression', 'Ridge_Regression','Lasso_Regression','Decision_Tree',
                    'Random_Forest','Neural_Network'], rotation=15)
ax.set_xlabel('Model')
ax.set_ylabel('MSE')
ax.legend(loc='upper left')
plt.show()
```

```python
import numpy as np
```

```python

species = ("k_NN", "Linear_Regression", "Ridge_Regression", "Lasso_Regression",
           "Decision_Tree","Random_Forest","Neural_Network")
penguin_means = {
    'MSE_train': (135259, 517659, 513183, 550842, 136980, 80370, 797153),
    'MSE_test': (139028, 91823, 109500, 148812, 216426, 153730, 325785),
}

x = np.arange(len(species))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in penguin_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Length (mm)')
ax.set_title('Penguin attributes by species')
ax.set_xticks(x + width, species)
ax.legend(loc='upper left', ncols=3)

plt.show()
```
