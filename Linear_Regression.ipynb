```python
import pandas as pd
```

```python
BID = pd.read_csv('./dataset/BID1.csv')
BID.head(5)
```

```python
from sklearn.model_selection import train_test_split
```

```python

X = BID[["Open"]]
X.head()
```

```python
y = BID["Close"]
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size = 0.3, random_state = 1)

print('Labels counts in y:', len(y))
print('Labels counts in y_train:', len(y_train))
print('Labels counts in y_test:', len(y_test))
```

```python
from sklearn.linear_model import LinearRegression
```

```python

lr = LinearRegression()

lr.fit(X_train.values.reshape(-1, 1), y_train)

y_pred_test = lr.predict(X_test.values.reshape(-1,1))
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python

mse = mean_squared_error(y_test, y_pred_test)

print(f'The test MSE is {mse}')

r2 = r2_score(y_test, y_pred_test)

print(f'The test R-squred score is {r2}')
```

```python
import matplotlib.pyplot as plt
```

```python

fig, ax = plt.subplots()
ax.scatter(X, y, label = 'data points')

ax.plot(X_test, y_pred_test, color='red', linewidth = 2, label = 'model predictions')
ax.set_xlabel('Open')
ax.set_ylabel('Close')
ax.legend()
```

```python

mse_train_list = []
mse_test_list = []
coef_list = []
intercept_list = []

cols = [['Open'],['High'],['Low'],['Open','High'],['Open','Low'],['High','Low'],['Open','High','Low']]

for i in cols:

    X = BID[i]
    
    y = BID["Close"]
   
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)

    lr = LinearRegression()

    lr.fit(X_train, y_train)
    
    y_train_pred = lr.predict(X_train)
    y_test_pred = lr.predict(X_test)
    
    mse_train = mean_squared_error(y_train, y_train_pred)
    mse_test = mean_squared_error(y_test, y_test_pred)
    
    mse_train_list.append(mse_train)
    mse_test_list.append(mse_test)
    
    coef_list.append(lr.coef_)
    intercept_list.append(lr.intercept_)
```

```python
print("Intercept\tCoeficients\tTrain MSE\tTest MSE")
for i in range(len(mse_train_list)):
    print(f'{intercept_list[i]}\t{coef_list[i]}\t{mse_train_list[i]}\t{mse_test_list[i]}')
```

## Model selection cho Ridge Regression dng phng php holdout

```python
X = BID[['Open','High','Low']]
y = BID[['Close']]

X_train_val, X_test, y_train_val, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)

X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size = 0.25, random_state = 1)
```

```python
from sklearn.linear_model import Ridge
```

```python

train_mse = []
val_mse = []
test_mse = []

# List of hyper-parameter values 
#alphas = [1e-8, 1e-6,1e-4, 1e-2, 0.0, 1e2, 1e4, 1e6, 1e8]
#alphas = [1e-2, 0.0, 1e2, 5e2, 1e3, 5e3, 1e4, 5e4, 1e5, 5e5, 1e6, 5e6, 1e7, 5e7, 1e8]
#alphas = [1e5, 5e5, 1e6, 5e6, 6e6, 7e6, 8e6, 9e6, 1e7, 5e7, 1e8]
#alphas = [8e6, 9e6, 9e6 + 2e5, 9e6 + 4e5, 9e6 + 6e5, 9e6 + 8e5, 1e7, 5e7, 1e8]
alphas = [1e2, 1e3, 1e4, 1e5, 1e6,1e7, 1e8, 1e9]

for alpha in alphas:
   
    ridge_reg = Ridge(alpha)

    ridge_reg.fit(X_train, y_train)
    
    y_train_pred = ridge_reg.predict(X_train)
    
    y_val_pred = ridge_reg.predict(X_val)

    y_test_pred = ridge_reg.predict(X_test)
    
    mse_train = mean_squared_error(y_train, y_train_pred) 
    mse_val = mean_squared_error(y_val, y_val_pred)
    mse_test = mean_squared_error(y_test, y_test_pred)
    
    train_mse.append(mse_train)
    val_mse.append(mse_val)
    test_mse.append(mse_test)
    
    print('alpha: {:7} | train mse: {:5} | val mse: {:5} | test mse: {:5}'.
          format(alpha, round(mse_train, 2), round(mse_val, 2), round(mse_test, 2)))
```

```python

best_alpha = 10000000

print(f'The best alpha (hyper-parameter) value, alpha = {best_alpha}')
```

```python
fig, ax = plt.subplots(figsize = (12, 8))

ax.plot(alphas, train_mse, 's--', label = 'Train error', color = 'blue', linewidth = 2)

ax.plot(alphas, val_mse, 's-', label = 'Val error',color ='green', linewidth = 2)

ax.plot(alphas, test_mse, 's-', label = 'Test error',color ='purple', linewidth = 2)

ax.axvline(best_alpha, 0, 0.5, color = 'r', label = f'best alpha = {best_alpha}')

ax.set_xlabel('Value of Alpha', fontsize = 15)
ax.set_ylabel('Mean Squared Error',fontsize = 15)
ax.legend(loc = 'upper left', fontsize = 16)
ax.set_title('Mean Squared Error', fontsize = 20)
ax.set_xscale('log')
plt.tight_layout()
plt.show();
```

```python
ridge_reg = Ridge(best_alpha)

ridge_reg.fit(X_train_val, y_train_val)

y_train_pred = ridge_reg.predict(X_train_val)

y_test_pred = ridge_reg.predict(X_test)

mse_train = mean_squared_error(y_train_val, y_train_pred)
mse_test = mean_squared_error(y_test, y_test_pred)

print('train mse: {:5} | test mse: {:5}'.
        format(round(mse_train, 2), round(mse_test, 2)))

r2 = r2_score(y_train_val, y_train_pred)
print(f'The train R-squred score is {r2}')

r2 = r2_score(y_test, y_test_pred)
print(f'The test R-squred score is {r2}')
```
